**PROPOSED SOLUTION:**

In order to collect necessary data, it is extracted from Twitter using a API through R Studio environment. Using keywords such as “Apple”,”Mac”,”MacBook” This collected data is raw and needs to cleaned and pre-processed for further analysis.  The collected tweets contains a lot of unnecessary elements such as website URLs, punctuations, stop-words and numbers.  Once data pre-processing is done through R-Studio, the cleaned data is used for classification. For this task, the ”Orange” tool is used. Multiple machine learning algorithms are implemented such as Neural Networks, Naïve-Bayes, and K-Nearest Neighbor Classifiers. This tool is chosen for this project as it is user-friendly and easy to analyse data followed by data mining. The below snapshot shows the workflow implemented in this tool for data processing.

Figure 5 : Orange Workflow  used for Classification 

To carry out analysis techniques, a training set is required. This dataset is used to build a prediction model to predict possible outputs of an unknown dataset. There are two outputs that are being predicted in this system; Positive words that is represented as 1 and negative words that is represented as 0. The training dataset used in this system is the previous year transactions of Apple products with defined positive and negative outputs. The cleaned dataset collected from Twitter is used as the Testing Set. By creating the prediction model and predicting outputs, the proportion of emotions from the extracted tweets is studied. This information is used to further create the model. The key words that are used more frequently related to the MacBook series is evaluated from the tweets as well. This can be represented through a WordCloud diagram generated through the R-Studio.  If the number of negative comments are higher than that of positive, it is understood that the product requires further iterations to increase sales.
